```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np


quartet = sns.load_dataset('anscombe')
#print(quartet['x'][0])
q = [quartet[i*11: i*11+11] for i in range(0, 4)]
#print(q[0].axes)
varx = [q[i].var(axis=0, ddof=1)[0] for i in range(0, 4)]
vary = [q[i].var(axis=0, ddof=1)[1] for i in range(0, 4)]
meanx = [q[i].mean(axis=0)[0] for i in range(0, 4)]
meany = [q[i].mean(axis=0)[1] for i in range(0, 4)]
cov = [q[i].cov()['x']['y'] for i in range(0, 4)]
corr = [q[i].corr('pearson')['x']['y'] for i in range(0, 4)]
w = [cov[i]/varx[i] for i in range(0, 4)]
b = [meany[i] - meanx[i]*w[i] for i in range(0, 4)]
print('Mean values of x and y: ', meanx, meany)
print('Var of x and y: ', varx, vary)
print("Perason\'s correlation coefficient: ", corr)
fig, axes = plt.subplots(2, 2)
for i in range(0, 2): 
  for j in range(0, 2):
    axes[i, j].scatter(data = q[2 * i + j], x = 'x', y = 'y')
    axes[i, j].plot([2, 19], [b + 2 * w[2 * i + j], b + 19 * w[2 * i + j]])

```

Возьмем лабортаорную по эффекту Майкельсона, где нужно было провести прямую в осях r^2 и n, где измерялись r(n) 

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

xs = np.array([0, 1, 2, 3, 4, 5, 6]).reshape(-1, 1)
ys = np.array([0, 5, 7, 8, 9, 10, 11.5]) ** 2

model = LinearRegression().fit(xs, ys)
ys_pred = np.array([model.predict(xs)]).reshape(-1, 1)
print('R^2 невязка равна:', model.score(xs, ys))
plt.scatter(xs, ys)
plt.plot(xs, ys_pred)
plt.show()
```
